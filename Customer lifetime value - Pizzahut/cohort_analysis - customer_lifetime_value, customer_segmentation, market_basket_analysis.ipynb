```python
df['yearmonth_transaction'] = df['TransactionDate'].dt.year.astype("str") + '-' + df['TransactionDate'].dt.month.astype("str")+ '-' +'1'
```

```python
df.info()
```

```python
df['yearmonth_transaction'] = pd.to_datetime(df['yearmonth_transaction'])
```

```python
df['cohort_yearmonth'] = df.groupby("CustomerID")['yearmonth_transaction'].transform("min")
```

```python
df['cohort_index'] = (df['yearmonth_transaction'].dt.year - df['cohort_yearmonth'].dt.year)*12 + (df['yearmonth_transaction'].dt.month - df['cohort_yearmonth'].dt.month) + 1
```

```python
df.tail(10)
```

```python
customer_table = df.groupby(['cohort_yearmonth','cohort_index'])['CustomerID'].nunique().reset_index()
```

```python
customer_table
```

```python
customer_table = customer_table.pivot(index="cohort_yearmonth",columns="cohort_index",values="CustomerID")
customer_table
```

```python
customer_table_percentage = customer_table.divide(customer_table.iloc[:,0],axis=0)
```

```python
customer_table_percentage
```

```python
customer_table_percentage = round(customer_table_percentage,2)
```

```python
customer_table_percentage
```

```python
retention_rate = np.mean([0.03,0.02,0.02,0.02,0.01,0.02,0.01,0.01,0.02,0.02,0.02,0.02,0.02,0.02,0.03])
```

```python
retention_rate
```

```python
churn_rate = 1- retention_rate
churn_rate
```

```python
df1 = df.copy()
```

```python
df1 = df1.groupby("CustomerID").agg({"BillID": lambda x: x.count(),"SalesAmount": lambda x: x.sum()}).reset_index()
```

```python
df1
```

```python
df1.rename(columns={"BillID":"total_transaction","SalesAmount":"revenue"},inplace=True)
```

```python
df1
```

gross profit is 65%

```python
df1['profit'] = df1['revenue']*0.65
```

```python
df1
```

```python
df1['AOV'] = df1['profit'] / df1['total_transaction']
```

```python
df1
```

```python
APF = df1['total_transaction'].sum() / len(df1)
APF
```

```python
APF
```

```python
ACL = 1/churn_rate
ACL
```

```python
ACL
```

```python
df1['CLV'] = df1['AOV']*APF*ACL
```

```python
df1
```

CAC is 300000

```python
df1['customer_segmentation'] = df1['CLV'].apply(lambda x: "unqualified customer" if x<300000 else("qualified customer"))
```

```python
q1_salesamount = df_mba['SalesAmount'].quantile(0.25)
q2_salesamount = df_mba['SalesAmount'].quantile(0.5)
q3_salesamount = df_mba['SalesAmount'].quantile(0.75)
```

```python
df_mba['sales_segmentation'] = df_mba['SalesAmount'].apply(lambda x: "1" if x<q1_salesamount else("2" if x>=q1_salesamount and x<q2_salesamount else("3" if x>=q2_salesamount and x<q3_salesamount else("4")) ))
```

```python
df_dummiesmba = pd.get_dummies(df_mba)
df_dummiesmba
```

```python
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frequent_item = apriori(df_dummiesmba,min_support=0.05,use_colnames=True)
```

```python
frequent_item
```

```python
rules = association_rules(frequent_item,metric="lift",min_threshold=1)
rules
```

```python
sliced_df = rules[rules['consequents'].apply(lambda x:("customer_segmentation_unqualified customer" in x and len(x)==1) or ("customer_segmentation_qualified customer" in x and len(x)==1))]
```

```python
sliced_df
```

```python
result = sliced_df[(sliced_df['lift'] > 1) & (sliced_df['confidence']>0.5) & (sliced_df['support']>0.2)]
```
