```python
df1_recency = df1.copy()
```

```python
df1_recency = df1_recency.groupby("CustomerID")["TransactionDate"].max().reset_index()
```

```python
df1_recency
```

```python
import numpy as np
```

```python
max_day = np.max(df1['TransactionDate'])
```

```python
df1_recency["datediff"] = df1_recency["TransactionDate"].apply(lambda x: max_day - x)
```

```python
df1_recency
```

```python
df1_recency['datediff'] = df1_recency['datediff'].dt.days
```

```python
df1_recency.sort_values(by=["datediff"],ascending=True)
```

```python
df1_recency.info()
```

```python
q1_recency = df1_recency['datediff'].quantile(0.25)
q2_recency = df1_recency['datediff'].quantile(0.5)
q3_recency = df1_recency['datediff'].quantile(0.75)
```

```python
print(q1_recency,q2_recency,q3_recency)
```

```python
df1_recency['R'] = df1_recency['datediff'].apply(lambda x: 4 if x < q1_recency else(3 if(x>=q1_recency) and (x<q2_recency) else(2 if(x>=q2_recency) and (x<q3_recency) else(1))))
```

```python
df1
```

```python
df1_frequency = df1.groupby("CustomerID")["BillID"].count().reset_index(name="frequency")
```

```python
df1_frequency
```

```python
q1_frequency = df1_frequency['frequency'].quantile(0.25)
q2_frequency = df1_frequency['frequency'].quantile(0.5)
q3_frequency = df1_frequency['frequency'].quantile(0.75)
```

```python
print(q1_frequency,q2_frequency,q3_frequency)
```

```python
df1_frequency.groupby("frequency")["CustomerID"].count().reset_index(name="count")
```

```python
df1_frequencyexclude1 = df1_frequency[df1_frequency["frequency"]!=1]
```

```python
q1_frequency = df1_frequencyexclude1['frequency'].quantile(0.25)
q2_frequency = df1_frequencyexclude1['frequency'].quantile(0.5)
q3_frequency = df1_frequencyexclude1['frequency'].quantile(0.75)
```

```python
print(q1_frequency,q2_frequency,q3_frequency)
```

```python
df1_frequency["F"] = df1_frequency['frequency'].apply(lambda x: 1 if x<q1_frequency else(2 if (x>=q1_frequency) and (x<q2_frequency) else(3 if (x>=q2_frequency) and (x<q3_frequency)
                                                                                                                                          else(4))))
```

```python
df1_frequency
```

```python
df1_frequency.groupby("F")['CustomerID'].count().reset_index()
```

```python
df1_monetary = df1.groupby("CustomerID")["SalesAmount"].sum().reset_index()
```

```python
df1_monetary
```

```python
q1_monetary = df1_monetary["SalesAmount"].quantile(0.25)
q2_monetary = df1_monetary["SalesAmount"].quantile(0.5)
q3_monetary = df1_monetary["SalesAmount"].quantile(0.75)
```

```python
print(q1_monetary,q2_monetary,q3_monetary)
```

```python
df1_monetary["M"] = df1_monetary["SalesAmount"].apply(lambda x: 1 if x<q1_monetary else(2 if (x>=q1_monetary) and (x<q2_monetary) else(3 if (x >=q2_monetary) and (x<q3_monetary) else(4))))
```

```python
df1_monetary
```

```python
df_rfm = pd.merge(pd.merge(df1_recency,df1_frequency,on="CustomerID"),df1_monetary,on="CustomerID")
```
