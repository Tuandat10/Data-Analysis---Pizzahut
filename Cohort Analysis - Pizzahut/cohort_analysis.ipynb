```python
import datetime as dt
```

```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'])
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['yearmonth_transaction'] = df['TransactionDate'].dt.year.astype("str") + '-' + df['TransactionDate'].dt.month.astype("str") + '-' + '1'
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['yearmonth_transaction'] = pd.to_datetime(df['yearmonth_transaction'])
```

```python
df.info()
```

```python
df.head(5)
```

```python
df['cohort_yearmonth'] = df.groupby("CustomerID")['yearmonth_transaction'].transform("min")
```

```python
df.tail(5)
```

```python
df.info()
```

```python
df['cohort_index'] = (df['yearmonth_transaction'].dt.year - df['cohort_yearmonth'].dt.year)*12 + (df['yearmonth_transaction'].dt.month - df['cohort_yearmonth'].dt.month) +1
```

```python
df.tail(5)
```

```python
cohort_customer = df.groupby(['cohort_yearmonth','cohort_index'])['CustomerID'].nunique().reset_index()
cohort_customer
```

```python
cohort_customer = cohort_customer.pivot(index="cohort_yearmonth",columns="cohort_index",values="CustomerID")
cohort_customer
```

```python
cohort_customer.index = cohort_customer.index.strftime("%B %Y")
```

```python
cohort_customer
```

```python
cohort_customerpercentage = cohort_customer.divide(cohort_customer.iloc[:,0],axis=0)
cohort_customerpercentage
```

```python
plt.figure(figsize=(21,10))
sns.heatmap(cohort_customerpercentage,annot=True,cmap="Blues")
```
